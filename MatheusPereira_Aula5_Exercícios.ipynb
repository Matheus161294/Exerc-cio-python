```
import math 

bia = t
bia.initializeTurtle()
bia.speed(10)

 
bia.penup()
bia.left(80)
bia.forward(200)
bia.right(90)
bia.pendown()

def figura(lados):
  linha = 100
  ang1 = 360/lados
  ang2 = (180 - ang1)/2
  raio = linha*(math.sin(ang2*math.pi/180)/math.sin(ang1*math.pi/180))

  for i in range(lados):
    bia.forward(int(raio))
    bia.left(180-ang2)
    bia.forward(linha)
    bia.left(180-ang2)
    bia.forward(int(raio))
    bia.left(180)

  bia.penup()
  bia.right(90)
  bia.forward(250)
  bia.left(90)
  bia.pendown()

figura(5)
figura(6)
figura(7)
```
